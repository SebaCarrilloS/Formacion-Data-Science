# 1. Optimice la siguiente función utilizando algún método numérico tal como , Gradient Descent, Newton Raphson , el de la secante , etc. Muestre que lo que encuentra es efectivamente un mínimo global.

$f(x,y)=(x-5)^{2}+(y-10)^{2}+16$

*  Para este problema usaremos Gradient Descent.

Se aprecia de la función dada que corresponde a una parábola de la forma:

$f(x,y)=a(x-h)^{2}+b(y-k)^{2}+c$

En este caso, se tiene que $h=5$ y $k=10$, en este caso, se tiene que $a=1$ y $b=1$, y al ser ambos un valor positivo, se tiene que la parábola se abre hacia arriba, lo que implica que de encontrar el mínimo local, se encontraría el mínimo global.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Definimos la ecuación de la circunferencia.
def f(x, y):
    return (x - 5)**2 + (y - 10)**2 + 16
```

Para el gradiente, tenemos que derivar respecto a $x$ y $y$. Estas derivadas son sencillas de calcular, por lo que el código lo anotamos a continuación como un código en numpy para facilitar su uso.

```python
def gradiente(x, y):
    return np.array([2 * (x - 5), 2 * (y - 10)])
```

Para el gradient descent, necesitamos de varias características, las cuales reflejaremos en una función:

*  `inicio`: Punto donde se iniciará la búsqueda.
*  `paso`: Corresponde a la tasa de aprendizaje que usaremos en la dirección opuesta a la gradiente.
*  `iter`: Cuántas iteraciones generaremos.

Con estos factores procederemos a construir la función:

```python
def gradient_descent(inicio,paso,iter):
  punto=np.array(inicio)
  camino=[punto.copy()]
  for _ in range(iter):
    grad=gradiente(punto[0], punto[1])
    punto=punto-paso*grad
    camino.append(punto.copy())

  return punto, camino
```

Con estas funciones, procedemos a definir los parámetros:

*  inicio= $(0,0)$
*  paso= $0,1$
*  iter= $1000$

```python
inicio=[0,0]
paso=0.1
iter=1000
```

```python
minimo, camino=gradient_descent(inicio,paso,iter)
```

Con esto, vemos los resultados finales.

```python
print(f"El mínimo encontrado es: {minimo}")
print(f"El valor de la función en el mínimo es: {f(minimo[0], minimo[1])}")
```

Dado que se encontró el mínimo global, y por lo expuesto en el inicio del problema, vemos que se encontró el mínimo global.

*  Dado que se ve que hay muchos outliers, se usará metodología de `IsolationForest` para la limpieza del DataFrame. Usaremos esta técnica dado que está diseñada específicamente para la detección de outliers, además de ser un modelo eficiente en conjuntos que poseen muchos datos, y también para ahorrar tiempo en este procedimiento.

*  Consideraremos que el 10% de los datos son outliers, por lo que daremos el parámetro de contaminación igual a $0.1$, además que este valor no es demasiado restrictivo.

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.1, random_state=42)

# Ajusta el modelo y predice los outliers
df['outlier'] = model.fit_predict(df)

# Filtra los datos para eliminar los outliers
df_limpio = df[df['outlier'] == 1].drop(columns=['outlier'])
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_limpio['precio'], kde=True, bins=30)
plt.title('Distribución del Precio de las Propiedades (datos limpios)')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=7, figsize=(28, 6), sharey=False)

# Tenemos que asegurarnos que cada eje (axe) sea un array para la correcta iteración
axes = axes.flatten()

# Generaremos un boxplot para cada columna
for i, column in enumerate(df.columns[:7]):
    sns.boxplot(y=df[column], ax=axes[i])
    axes[i].set_title(column)
    axes[i].set_xlabel('')

plt.tight_layout()
```

# 3. Simule un modelo AR(2) con $10.000$ datos de la siguiente manera . Realice esto $1.000$ veces (es decir genere una serie de $10.000$ datos del proceso autorregresivo $1.000$ veces)

 $y_{t}=0.5+0.3y_{t-1}+0.2y_{t-2}+\epsilon_{t}$


Donde $y_{t}$ es la temperatura en grados centígrados y $\epsilon_{t}$ es un shock normal con media cero y desviación estándar 5. Suponga que $y_{0} = 20$ y $\epsilon_{0}=0$ .

a) Grafique la serie simulada.


```python
## Llamamos a las librerías que usaremos en los problemas:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.ar_model import AutoReg
```

Procedemos a definir parámetros de la serie temporal:

*  $n=10000$: Longitud de cada serie
*  Numero de series = $1000$
*  $\phi_{0}=0.5$ : Intercepto
*  $\phi_{1}=0.3$ : Coeficiente para $y_{1}$
*  $\phi_{2}=0.2$ :  Coeficiente para $y_{2}$
*  $y_{0}=20$_:  Valor inical para $y_{1}$
*  $\sigma=5$ : Desviación estándar


```python
n = 10000
num_series = 1000
phi_0 = 0.5
phi_1 = 0.3
phi_2 = 0.2
y_0 = 20
std_dev = 5

# Inicializar una lista para almacenar las series simuladas
series_simuladas = []

for _ in range(num_series):
    y = np.zeros(n)
    y[0] = y_0
    epsilon = np.random.normal(0, std_dev, n)

    # Ahora pasamos a simular el modelo AR(2)
    for t in range(2, n):
        y[t] = phi_0 + phi_1 * y[t - 1] + phi_2 * y[t - 2] + epsilon[t]

    # Con estos datos, pasamos a agregar la serie simulada a la lista previamente hecha.
    series_simuladas.append(y)

# Convertimos en DataFrame
df_series = pd.DataFrame(series_simuladas).T
```

```python
serie_a_graficar = df_series[0]

# Generamos gráfico
plt.figure(figsize=(12, 6))
plt.plot(serie_a_graficar, color='blue', linewidth=1)
plt.title("Serie Simulada de Temperatura (AR(2))")
plt.xlabel("Tiempo")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
```

b) Muestre la ACF y la PACF y explique si esta es coherente con el proceso simulado de acuerdo con el proceso de Box y Jenkins .


**Respuesta:**

Para el correcto análisis solicitado en la pregunta, se procederá con los siguientes pasos, los cuales son parte del enfoque de Box y Jenkins para la identificación del proceso:

1.  ACF (Autocorrelación): Este mide la correlación que existe entre una variable y sus retardos. Para este proceso en particular ($AR(2)$), se tiene que la ACF debiese disminuir de manera lenta a medida que los retardos aumentan. Esto sucede gracias a que la dependencia que tiene persiste hasta 2 retardos.
2.  PACF (Autocorrelación Parcial): Esta muestra la correlación entre la variable y sus retardos después de haber eliminado el efecto de los retardos intermedios. En el proceso de $AR(2)$, se debiese tener que la PACF mostrase una caída significativa después de los primeros 2 retardos, y esto se debe a la dependencia que tiene de $y_{t-1}$ y $y_{t-2}$, pero no de retardos mayores.

```python
# Seleccionamos la primera serie simulada
serie_a_analizar = df_series[0]

# Tenemos que graficar la ACF y la PACF
# Para ello, generaremos los 2 gráficos haciendo subplots
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# ACF
plot_acf(serie_a_analizar, lags=20, ax=ax[0])
ax[0].set_title('Función de Autocorrelación (ACF)')

# PACF
plot_pacf(serie_a_analizar, lags=20, ax=ax[1], method='ywm')
ax[1].set_title('Función de Autocorrelación Parcial (PACF)')

plt.tight_layout()
```

Analizando el gráfico, se puede observar que en el caso del ACF la función de autocorrelación muestra como disminuye la correlación a medida que aumentan los retardos, y la disminución es lenta, lo cual es coherente con lo esperado para un proceso $AR(2)$

Para la PACF se logra apreciar que los 2 primeros retardos son importantes, y después hay un corte abrupto, con lo que el PACF cae rápidamente, lo que según  Box y Jenkins es lo esperado en un proceso $AR(2)$.

Grafique la distribución para la constante de su modelo y el parámetro del AR( 1 )). ¿ Qué observa en la distribución de los p ar ámetros estimados. Explique a qué se deben estos resultados.

En este caso, tendremos que suponer un modelo $AR(1)$, por lo cual, generaremos los siguientes pasos:

1.  Estimar el modelo $AR(1)$. Para ello usaremos la forma:

 $y_{t}=c+\phi_{1}y_{t-1}+\epsilon_{t}$

 Donde se tiene que:
 *  $\phi_{1}$ es el parámetro del modelo $AR(1)$
 *  $c$ es una constante.


2.  Después de ajustar el modelo $AR(1)$ a las series, se tendrá que ajustar tanto $c$ como $\phi_{1}$.

```python
constante = []
ar1_coefficients = []

# Ajustar un modelo AR(1) para cada una de las 1000 series simuladas
for series in df_series.T.values:
    model = AutoReg(series, lags=1).fit()
    constante.append(model.params[0])  # Constante estimada
    ar1_coefficients.append(model.params[1])  # Coeficiente AR(1) estimado
```

```python
# Graficar las distribuciones de los parámetros estimados
plt.figure(figsize=(12, 6))

# Distribución de la constante
plt.subplot(1, 2, 1)
plt.hist(constante, bins=30, color='b', edgecolor='black', alpha=0.7)
plt.title('Distribución de la constante (c) estimada')
plt.xlabel('Constante estimada')
plt.ylabel('Frecuencia')

# Distribución del coeficiente AR(1)
plt.subplot(1, 2, 2)
plt.hist(ar1_coefficients, bins=30, color='r', edgecolor='black', alpha=0.7)
plt.title('Distribución del coeficiente AR(1) estimado')
plt.xlabel('Coeficiente AR(1) estimado')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

De los gráficos expuestos, se ve como $c$ toma valores entre $0.5$ y $0.75$, y este valor asigna un valor positivo a la temperatura base, lo cual tiene sentido. Sin embargo, se ve en el coeficiente $\phi_{1}$, que va entre $0.35$ y $0.4$, siendo este valor sobre-estimado, ya que trata de captar la dinámica que posee un modelo $AR(2)$ y el valor debiese ser menor al obtenido. Es decir, el modelo de $AR(1)$ no logra captar la influencia que está teniendo $y_{2}$ en la fórmula original, que corresponde al valor de $y_{t}$

Esto muestra que el modelo $AR(1)$ no es el adecuado para este problema.
