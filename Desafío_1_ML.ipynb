Dado que estos casos son solo 2 para ambos casos (1 habitación y 6 habitaciones) y estos son muy, los eliminaremos del dataset.

```python
df=df.drop(df[(df["bedrooms"]==6) | (df["bedrooms"]==1)].index)
```

```python
#Similar con los precios, buscamos precios fuera de rango:
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar los datos eliminando los outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Aplicar la función para eliminar outliers en la columna 'price'
df = remove_outliers(df, 'price')
```

```python
plt.figure(figsize=(10,6),dpi=150)
sns.heatmap(df.corr(), annot=True)
```

## 8. Genera un modelo de regresión lineal. Para ello:

a. Importa el estimador de regresión lineal de Scikit-learn.

b. Divide los datos (80% train y 20% test)

c. Crea una instancia del modelo de regresión lineal.

d. Ajusta el modelo utilizando los datos de entrenamiento.

```python
X=df.drop("price",axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

9. Evalúa el modelo. Para ello:

a. Utiliza los datos de entrenamiento y calcula métricas de evaluación, como el error cuadrado medio (MSE) o el coeficiente de determinación $R^{2}$. Grafica los coeficientes.

b. Utiliza los datos de prueba para realizar predicciones con el modelo
entrenado, y evalúa el rendimiento del modelo comparando las predicciones con los valores reales.

```python
coeficientes = model.coef_
df_coeficientes = pd.DataFrame({
    'Variable': X_train.columns,
    'Coeficiente': coeficientes
})
df_coeficientes = df_coeficientes.reindex(df_coeficientes['Coeficiente'].abs().sort_values(ascending=False).index)
```

```python
df_coeficientes
```

```python
plt.figure(figsize=(10, 6));
sns.barplot(x='Coeficiente', y='Variable', data=df_coeficientes, hue='Variable', palette='viridis', legend=False);
plt.title('Coeficientes del Modelo de Regresión Lineal');
plt.xlabel('Valor del Coeficiente');
plt.ylabel('Variable');
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.round(np.sqrt(mse),2)

mae=np.round(mean_absolute_error(y_test,y_pred),2)
```

```python
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```
